Realizamos la misma operación para el segundo conjunto formateando y limpiando columnas acorde a lo requerido

```python

df_SAP = pd.read_csv('/content/SONY_Exercise_1 - SAP.csv',sep=',',header=0)
df_SAP.rename(columns={'    Net': 'Saldo','Customer nr.':'CustomerNR'}, inplace=True)
df_SAP['Saldo'].astype(str)
df_SAP['Saldo'] = df_SAP['Saldo'].str.replace(',', '.')
df_SAP['Saldo'] = df_SAP['Saldo'].astype(float)
#tipos_de_datos = df_SAP.dtypes
#print(tipos_de_datos)
#print(df_SAP.columns) -> nombre columnas
df_fixed_SAP = df_SAP.groupby(['CustomerNR','Customer name']).agg({'Saldo':'sum','Processed quantity':'sum'}).reset_index().sort_values(['Processed quantity'],ascending=False).copy()
df_fixed_SAP.head(13)
```

Se uniran ambas tablas por el tipo de cliente y se validara si la cantidad procesada corresponde al stock por material

```python
print(df_customer.columns)
df_customer.reset_index(drop=True, inplace=True)
merged_df= pd.merge(df_fixed_SAP[['CustomerNR','Customer name','Saldo','Processed quantity']],df_customer[['CustomerNR', 'QTY to compensate']], on='CustomerNR', how='inner')
merged_df['Coincide'] = merged_df.apply(lambda row: True if row['Processed quantity'] == row['QTY to compensate'] else False, axis=1)
merged_df.head(12)

```
